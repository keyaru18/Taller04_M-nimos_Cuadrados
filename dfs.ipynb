```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
xs = [0.0003, 0.0822, 0.2770, 0.4212, 0.4403, 0.5588, 0.5943, 0.6134, 0.9070, 1.0367,
      1.1903, 1.2511, 1.2519, 1.2576, 1.6165, 1.6761, 2.0114, 2.0557, 2.1610, 2.6344]
ys = [1.1017, 1.5021, 0.3844, 1.3251, 1.7206, 1.9453, 0.3894, 0.3328, 1.2887, 3.1239,
      2.1778, 3.1078, 4.1856, 3.3640, 6.0330, 5.8088, 10.5890, 11.5865, 11.8221, 26.5077]

x = np.array(xs)
y = np.array(ys)

# ==============================================
# 1. FUNCI√ìN PARA AJUSTE LINEAL (GRADO 1)
# ==============================================
def ajuste_lineal():
    print("\n=== AJUSTE LINEAL (GRADO 1) ===")

    # Construir matriz X para grado 1
    X1 = np.column_stack([np.ones(len(x)), x])

    # Calcular (X·µÄX) y (X·µÄy)
    XT_X1 = X1.T @ X1
    XT_y1 = X1.T @ y

    # Resolver sistema
    theta1 = np.linalg.solve(XT_X1, XT_y1)

    # Resultados
    print(f"\nEcuaci√≥n: y = {theta1[1]:.4f}x + {theta1[0]:.4f}")

    # Gr√°fico
    plt.figure(figsize=(10, 5))
    plt.scatter(x, y, color='red', label='Datos')
    x_line = np.linspace(min(x), max(x), 100)
    y_line = theta1[1]*x_line + theta1[0]
    plt.plot(x_line, y_line, 'b-', label=f'Regresi√≥n lineal')
    plt.title("Ajuste Lineal (Grado 1)")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)
    plt.show()

# ==============================================
# 2. FUNCI√ìN PARA AJUSTE CUADR√ÅTICO (GRADO 2)
# ==============================================
def ajuste_cuadratico():
    print("\n=== AJUSTE CUADR√ÅTICO (GRADO 2) ===")

    # Construir matriz X para grado 2
    X2 = np.column_stack([np.ones(len(x)), x, x**2])

    # Calcular (X·µÄX) y (X·µÄy)
    XT_X2 = X2.T @ X2
    XT_y2 = X2.T @ y

    # Resolver sistema
    theta2 = np.linalg.solve(XT_X2, XT_y2)

    # Resultados
    print(f"\nEcuaci√≥n: y = {theta2[2]:.4f}x¬≤ + {theta2[1]:.4f}x + {theta2[0]:.4f}")

    # Gr√°fico
    plt.figure(figsize=(10, 5))
    plt.scatter(x, y, color='red', label='Datos')
    x_curve = np.linspace(min(x), max(x), 100)
    y_curve = theta2[2]*x_curve**2 + theta2[1]*x_curve + theta2[0]
    plt.plot(x_curve, y_curve, 'g-', label='Regresi√≥n cuadr√°tica')
    plt.title("Ajuste Cuadr√°tico (Grado 2)")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)
    plt.show()

# Llamar a las funciones para mostrar ambos ajustes
ajuste_lineal()
ajuste_cuadratico()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
xs = [0.0003, 0.0822, 0.2770, 0.4212, 0.4403, 0.5588, 0.5943, 0.6134, 0.9070, 1.0367,
      1.1903, 1.2511, 1.2519, 1.2576, 1.6165, 1.6761, 2.0114, 2.0557, 2.1610, 2.6344]
ys = [1.1017, 1.5021, 0.3844, 1.3251, 1.7206, 1.9453, 0.3894, 0.3328, 1.2887, 3.1239,
      2.1778, 3.1078, 4.1856, 3.3640, 6.0330, 5.8088, 10.5890, 11.5865, 11.8221, 26.5077]

x = np.array(xs)
y = np.array(ys)

# ==============================================
# 1. FUNCI√ìN PARA AJUSTE LINEAL (GRADO 1)
# ==============================================
def ajuste_lineal():
    print("\n=== AJUSTE LINEAL (GRADO 1) ===")
    print("\nFunci√≥n objetivo a minimizar:")
    print("S(Œ∏‚ÇÄ, Œ∏‚ÇÅ) = Œ£(y·µ¢ - (Œ∏‚ÇÄ + Œ∏‚ÇÅx·µ¢))¬≤")
    
    # Construir matriz X para grado 1
    X1 = np.column_stack([np.ones(len(x)), x])
    
    # Mostrar sistema de ecuaciones normales
    print("\nSistema de ecuaciones normales:")
    print("nŒ∏‚ÇÄ + Œ∏‚ÇÅŒ£x·µ¢ = Œ£y·µ¢")
    print("Œ∏‚ÇÄŒ£x·µ¢ + Œ∏‚ÇÅŒ£x·µ¢¬≤ = Œ£x·µ¢y·µ¢")
    
    # Calcular sumatorias para mostrar
    n = len(x)
    sum_x = sum(x)
    sum_y = sum(y)
    sum_x2 = sum(x**2)
    sum_xy = sum(x*y)
    
    print("\nValores calculados:")
    print(f"n = {n}")
    print(f"Œ£x·µ¢ = {sum_x:.4f}")
    print(f"Œ£y·µ¢ = {sum_y:.4f}")
    print(f"Œ£x·µ¢¬≤ = {sum_x2:.4f}")
    print(f"Œ£x·µ¢y·µ¢ = {sum_xy:.4f}")
    
    print("\nSistema num√©rico:")
    print(f"{n}Œ∏‚ÇÄ + {sum_x:.4f}Œ∏‚ÇÅ = {sum_y:.4f}")
    print(f"{sum_x:.4f}Œ∏‚ÇÄ + {sum_x2:.4f}Œ∏‚ÇÅ = {sum_xy:.4f}")

    # Calcular (X·µÄX) y (X·µÄy)
    XT_X1 = X1.T @ X1
    XT_y1 = X1.T @ y

    # Resolver sistema
    theta1 = np.linalg.solve(XT_X1, XT_y1)

    # Resultados
    print(f"\nSoluci√≥n del sistema:")
    print(f"Œ∏‚ÇÄ = {theta1[0]:.4f}")
    print(f"Œ∏‚ÇÅ = {theta1[1]:.4f}")
    print(f"\nEcuaci√≥n final: y = {theta1[1]:.4f}x + {theta1[0]:.4f}")

    # Gr√°fico
    plt.figure(figsize=(10, 5))
    plt.scatter(x, y, color='red', label='Datos')
    x_line = np.linspace(min(x), max(x), 100)
    y_line = theta1[1]*x_line + theta1[0]
    plt.plot(x_line, y_line, 'b-', label=f'Regresi√≥n lineal')
    plt.title("Ajuste Lineal (Grado 1)")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)
    plt.show()

# ==============================================
# 2. FUNCI√ìN PARA AJUSTE CUADR√ÅTICO (GRADO 2)
# ==============================================
def ajuste_cuadratico():
    print("\n=== AJUSTE CUADR√ÅTICO (GRADO 2) ===")
    print("\nFunci√≥n objetivo a minimizar:")
    print("S(Œ∏‚ÇÄ, Œ∏‚ÇÅ, Œ∏‚ÇÇ) = Œ£(y·µ¢ - (Œ∏‚ÇÄ + Œ∏‚ÇÅx·µ¢ + Œ∏‚ÇÇx·µ¢¬≤))¬≤")
    
    # Construir matriz X para grado 2
    X2 = np.column_stack([np.ones(len(x)), x, x**2])
    
    # Mostrar sistema de ecuaciones normales
    print("\nSistema de ecuaciones normales:")
    print("nŒ∏‚ÇÄ + Œ∏‚ÇÅŒ£x·µ¢ + Œ∏‚ÇÇŒ£x·µ¢¬≤ = Œ£y·µ¢")
    print("Œ∏‚ÇÄŒ£x·µ¢ + Œ∏‚ÇÅŒ£x·µ¢¬≤ + Œ∏‚ÇÇŒ£x·µ¢¬≥ = Œ£x·µ¢y·µ¢")
    print("Œ∏‚ÇÄŒ£x·µ¢¬≤ + Œ∏‚ÇÅŒ£x·µ¢¬≥ + Œ∏‚ÇÇŒ£x·µ¢‚Å¥ = Œ£x·µ¢¬≤y·µ¢")
    
    # Calcular sumatorias para mostrar
    n = len(x)
    sum_x = sum(x)
    sum_y = sum(y)
    sum_x2 = sum(x**2)
    sum_x3 = sum(x**3)
    sum_x4 = sum(x**4)
    sum_xy = sum(x*y)
    sum_x2y = sum(x**2 * y)
    
    print("\nValores calculados:")
    print(f"n = {n}")
    print(f"Œ£x·µ¢ = {sum_x:.4f}")
    print(f"Œ£y·µ¢ = {sum_y:.4f}")
    print(f"Œ£x·µ¢¬≤ = {sum_x2:.4f}")
    print(f"Œ£x·µ¢¬≥ = {sum_x3:.4f}")
    print(f"Œ£x·µ¢‚Å¥ = {sum_x4:.4f}")
    print(f"Œ£x·µ¢y·µ¢ = {sum_xy:.4f}")
    print(f"Œ£x·µ¢¬≤y·µ¢ = {sum_x2y:.4f}")
    
    print("\nSistema num√©rico:")
    print(f"{n}Œ∏‚ÇÄ + {sum_x:.4f}Œ∏‚ÇÅ + {sum_x2:.4f}Œ∏‚ÇÇ = {sum_y:.4f}")
    print(f"{sum_x:.4f}Œ∏‚ÇÄ + {sum_x2:.4f}Œ∏‚ÇÅ + {sum_x3:.4f}Œ∏‚ÇÇ = {sum_xy:.4f}")
    print(f"{sum_x2:.4f}Œ∏‚ÇÄ + {sum_x3:.4f}Œ∏‚ÇÅ + {sum_x4:.4f}Œ∏‚ÇÇ = {sum_x2y:.4f}")

    # Calcular (X·µÄX) y (X·µÄy)
    XT_X2 = X2.T @ X2
    XT_y2 = X2.T @ y

    # Resolver sistema
    theta2 = np.linalg.solve(XT_X2, XT_y2)

    # Resultados
    print(f"\nSoluci√≥n del sistema:")
    print(f"Œ∏‚ÇÄ = {theta2[0]:.4f}")
    print(f"Œ∏‚ÇÅ = {theta2[1]:.4f}")
    print(f"Œ∏‚ÇÇ = {theta2[2]:.4f}")
    print(f"\nEcuaci√≥n final: y = {theta2[2]:.4f}x¬≤ + {theta2[1]:.4f}x + {theta2[0]:.4f}")

    # Gr√°fico
    plt.figure(figsize=(10, 5))
    plt.scatter(x, y, color='red', label='Datos')
    x_curve = np.linspace(min(x), max(x), 100)
    y_curve = theta2[2]*x_curve**2 + theta2[1]*x_curve + theta2[0]
    plt.plot(x_curve, y_curve, 'g-', label='Regresi√≥n cuadr√°tica')
    plt.title("Ajuste Cuadr√°tico (Grado 2)")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)
    plt.show()

# Llamar a las funciones para mostrar ambos ajustes
ajuste_lineal()
ajuste_cuadratico()
```
